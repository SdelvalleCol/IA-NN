```python
import wfdb
from numpy import var , std
import matplotlib.pyplot as plt
pepito = wfdb.rdrecord("database/103")
p_singla = pepito.p_signal

i_inicial = 0 
i_final = 7
data_final = []
while(True):
  try:
    data = []
    for i in range(i_inicial,i_final):
      data.append(p_singla[i][0])
    varianza = var(data)
    desviacion = std(data)
    for i in range(0,7):
      data_final.append(desviacion)
    i_inicial += 7
    i_final += 7
  except:
    longitud_dif = len(p_singla)-len(data_final)
    for i in range(0,longitud_dif):
      data_final.append(0)
    break

print(len(p_singla))
print(len(data_final))
plt.plot(data_final[:1000])
##Cambiar rango si se desea

  
  
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from scipy import stats
import numpy as np
import pandas as pd
```

```python
data_c = pd.read_csv('cancer_mama.csv')
print(data_c)
```

```python
del data_c ["842302"]
print(data_c)
```

```python
data_c = data_c.rename({
'M': 'CS'
, '17.99': 'A'
, '10.38': 'B'
, '122.8': 'C'
, '1001': 'D'
, '0.1184': 'E'
, '0.2776': 'F'
, '0.3001': 'G'
, '0.1471': 'H'
, '0.2419': 'I'
, '25.38': 'J'
, '17.33': 'K'
, '184.6': 'M'
, '2019': 'N'
, '0.1622': 'Ñ'
, '0.6656': 'O'
, '0.7119': 'P'
, '0.2654': 'Q'
, '0.4601': 'R'
, '0.1189': 'S'
}, axis=1)

print(data_c)
```

```python
data_c= data_c.reindex(columns=["A","B","C","D","E","F","G","H","I","J","K","M","N","Ñ","O","P","Q","R","S","CS"])
print(data_c)
```

```python
data_c['CS'] = data_c['CS'].replace(['M', 'B'], [1, 0])
print(data_c)
```

```python
data_c.boxplot(column=["A","B","C","D","E","F","G","H","I","J","K","M","N","Ñ","O","P","Q","R","S","CS"])
print(data_c)
```

```python
import numpy as np
columnas = ["A","B","C","D","E","F","G","H","I","J","K","M","N","Ñ","O","P","Q","R","S"]
for i in columnas:
  superior = np.quantile(data_c[i],.75)
  inferior = np.quantile(data_c[i],.25)
  iqr = superior -inferior
  limite_inferior = inferior - 1.5 * iqr
  limite_superior = superior + 1.5 * iqr
  data_c = data_c[data_c[i]<=limite_superior]
  data_c = data_c[data_c[i]>=limite_inferior]
print(len(data_c))
data_c.boxplot(column=["A","B","C","D","E","F","G","H","I","J","K","M","N","Ñ","O","P","Q","R","S","CS"])
```

```python
matrix =data_c.corr()
print(matrix)
```

```python
x=data_c[[ "A","B","C","D","E","F","G","H","I","J","K","M","N","Ñ","O","P","Q","R","S"]]
y=data_c[["CS"]]
y=y.astype('int')
X_train, X_test, y_train, y_test = train_test_split(x, y,train_size=0.3)
print(len(X_train))
clf = MLPClassifier(max_iter=200).fit(X_train, y_train)
clf.score(X_test, y_test) 

```
